```python
import pandas as pd
import numpy as np
```

```python
# read Orders sheet from the input file
df=pd.read_excel('Input_Sample - Superstore.xls',sheet_name='Orders')
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.head()
```

```python
# add order year 
df['year']=df['Order Date'].dt.year 
df.head()
```

```python
# find the first purchase year for each custom
# Here a cohort means a group that share the same 'first-purchase-year'
s=df.groupby(by='Customer Name').year.min()
s.values[0:10]
```

```python
# create dataframe contain all custom names with ID
df_name=df.loc[:,['Customer ID', 'Customer Name']]
df_name.drop_duplicates(inplace=True, ignore_index=True)
df_name['key']=0
df_name.sort_values(by='Customer Name',inplace=True, ignore_index=True)
df_name['First Purchase Date']=s.values
df_name
```

```python
# create dataframe with year as column
df_year=pd.DataFrame([[0,2018],[0,2019],[0,2020] ,[0,2021]], columns=['key','Year'])
df_year

```

```python
# join df_name with df_year so that each custom includes a year duration from 2018-2021
df_temp=df_name.merge(df_year, on='key', how='left')
df_temp.drop(columns='key',inplace=True)
df_temp.head()
```

```python
# join df_temp with df so that each custom info contains a year from 2018-2021 if there is no order in a year, then NAN
df_final=df.merge(df_temp, left_on=['Customer ID','Customer Name','year'], right_on=['Customer ID','Customer Name','Year'], how='right')
```

```python
# drop column year
df_final.drop(columns='year',inplace=True)
```

```python
df_final.head(10)
```

```python
# find  rows that having initial date <= Year
index=df_final['First Purchase Date']<=df_final['Year']
```

```python
# new dataframe only contains these index-rows
df_final1=df_final[index]
df_final1.shape
```

```python
#Convert columns to best possible dtypes using dtypes supporting pd.NA.
df_final1=df_final1.convert_dtypes() 
```

```python
# add column order? if a custom has made an order this year, its value is 1 otherwise 0
df_final1['Order?']=df_final1['Sales'].notna().astype(int)
```

```python
df_final1.head()
```

```python
# -- add column 'Last Year Order?'
# Generate a new dataframe S
S=df_final1.loc[:,['Customer Name','Year','Order?']].drop_duplicates()
S
```

```python
# -- add column 'Last Year Order?' to S
S['Last Year Order?']=S.groupby(by='Customer Name').shift(periods=1).convert_dtypes()['Order?']
S.drop(columns='Order?',inplace=True)
S
```

```python
#add 'last year order' to df_final1
df_final1=df_final1.merge(S, on=['Customer Name','Year'])
```

```python
# -- calculate Customer Classification
# new
df_final1.loc[df_final1['First Purchase Date']==df_final1['Year'],'Customer Classification']='New'
df_final1.head()
```

```python
# Sleeping
df_final1.loc[df_final1['Order?']==0,'Customer Classification']='Sleeping'
df_final1.head()
```

```python
# consistent
df_final1.loc[(df_final1['Last Year Order?']==1) & (df_final1['Order?']==1),'Customer Classification']='Consistent'
```

```python
pd.options.display.max_rows=None
```

```python
df_final1.head()
```

```python
# Returning
df_final1.loc[(df_final1['Last Year Order?']==0) & (df_final1['Order?']==1),'Customer Classification']='Returning'
# check if 'Customer Classification' has null 
df_final1['Customer Classification'].isna().any()
```

```python
df_final1.head(40)
```

```python
# calculate YoY Difference
# create df_unique to remove unnecessary and duplicated information from df_final1
df_unique=df_final1.loc[:,['Customer ID','Customer Name','First Purchase Date','Year','Order?','Last Year Order?','Customer Classification']]
df_unique.drop_duplicates(inplace=True)
```

```python
df_unique.shape
```

```python
# calculate number of active customers in 2018 with first_purchase_date=2018
Num_2018_2018=df_unique.loc[(df_unique['First Purchase Date']==2018)&(df_unique['Customer Classification']=='New'),'Order?'].sum()
Num_2018_2018
```

```python
# calculate number of active customers in 2019 with first_purchase_date=2018
Num_2018_2019=df_unique.loc[(df_unique['First Purchase Date']==2018)&(df_unique['Year']==2019),'Order?'].sum()
Num_2018_2019
```

```python
# calculate number of active customers in 2020 with first_purchase_date=2018
Num_2018_2020=df_unique.loc[(df_unique['First Purchase Date']==2018)&(df_unique['Year']==2020),'Order?'].sum()
Num_2018_2020
```

```python
# calculate number of active customers in 2021 with first_purchase_date=2018
Num_2018_2021=df_unique.loc[(df_unique['First Purchase Date']==2018)&(df_unique['Year']==2021),'Order?'].sum()
Num_2018_2021
```

```python
# calculate 'YoY Difference' for customers with first_purchase_date=2018
df_unique.loc[(df_unique['First Purchase Date']==2018)&(df_unique['Year']==2019),'YoY Difference']=Num_2018_2019-Num_2018_2018
df_unique.loc[(df_unique['First Purchase Date']==2018)&(df_unique['Year']==2020),'YoY Difference']=Num_2018_2020-Num_2018_2019
df_unique.loc[(df_unique['First Purchase Date']==2018)&(df_unique['Year']==2021),'YoY Difference']=Num_2018_2021-Num_2018_2020
df_unique.head()
```

```python
# calculate number of active customers in 2019 with first_purchase_date=2019
Num_2019_2019=df_unique.loc[(df_unique['First Purchase Date']==2019)&(df_unique['Customer Classification']=='New'),'Order?'].sum()
Num_2019_2019
```

```python
# calculate number of active customers in 2020 with first_purchase_date=2019
Num_2019_2020=df_unique.loc[(df_unique['First Purchase Date']==2019)&(df_unique['Year']==2020),'Order?'].sum()
Num_2019_2020
```

```python
# calculate number of active customers in 2021 with first_purchase_date=2019
Num_2019_2021=df_unique.loc[(df_unique['First Purchase Date']==2019)&(df_unique['Year']==2021),'Order?'].sum()
Num_2019_2021
```

```python
# calculate 'YoY Difference' for customers with first_purchase_date=2019
df_unique.loc[(df_unique['First Purchase Date']==2019)&(df_unique['Year']==2020),'YoY Difference']=Num_2019_2020-Num_2019_2019
df_unique.loc[(df_unique['First Purchase Date']==2019)&(df_unique['Year']==2021),'YoY Difference']=Num_2019_2021-Num_2019_2020
df_unique.head(40)
```

```python
# calculate number of active customers in 2020 with first_purchase_date=2020
Num_2020_2020=df_unique.loc[(df_unique['First Purchase Date']==2020)&(df_unique['Year']==2020),'Order?'].sum()
Num_2020_2020
```

```python
# calculate number of active customers in 2021 with first_purchase_date=2020
Num_2020_2021=df_unique.loc[(df_unique['First Purchase Date']==2020)&(df_unique['Year']==2021),'Order?'].sum()
Num_2020_2021
```

```python
# calculate 'YoY Difference' for customers with first_purchase_date=2020
df_unique.loc[(df_unique['First Purchase Date']==2020)&(df_unique['Year']==2021),'YoY Difference']=Num_2020_2021-Num_2020_2020
df_unique.head(40)
```

```python
# save df_unique to excel
df_unique.to_excel('Output_forTableau.xlsx', index=False)
```

```python
# generate required output dataframe by using df_unique and df_final1
df_final1=df_final1.merge(df_unique, on=['Customer ID','Customer Name','First Purchase Date','Year','Order?','Last Year Order?','Customer Classification'])
```
